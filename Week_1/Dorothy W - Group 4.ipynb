PART C: Data Manipulation (Use Functions, Loops, and Conditional Statements)

```python
# Question 1: Write a function calculate_annual_income() that takes Monthly_Income as input and returns Annual_Income. Apply it to the dataset.
# Define the function
def calculate_annual_income(Monthly_Income):
    return Monthly_Income * 12

# applying the function to the data - creating annual income column/field
df['Annual_Income'] = df['Monthly_Income'].apply(calculate_annual_income)
df.head(3)
```

```python
# Categorize Age into groups using an if-else statement and add a new column Age_Group:
   # 18-30: 'Young Adult'
    # 31-50: 'Middle Age'
    # 51-80: 'Senior Citizen'
# define a function to define a function to categorize age
def categorize_age(age):
    if 18 <= age <= 30:
        return 'Young Adult'
    elif 31 <= age <= 50:
        return 'Middle Age'
    elif 51 <= age <= 80:
        return 'Senior Citizen'
    else:
        return 'Other'
 # Creating a new column Age_Group and using the function
df['Age_Group'] = df['Age'].apply(categorize_age)
df.head(3)
```

```python
# •	Create a function categorize_transactions(num_transactions) that returns:
     #'Frequent' if Number_of_Transactions > 30
    #'Moderate' if 10 ≤ Number_of_Transactions ≤ 30
    # 'Low' if Number_of_Transactions < 10
# define a function to define a function to categorize_transactions
def categorize_transactions(num_transactions):
    if num_transactions > 30:
        return 'Frequent'
    elif 10 <= num_transactions <= 30:
        return 'Moderate'
    elif num_transactions < 10:
        return 'Low'
    else:
        return 'Other'
 # Use a for loop to apply this function to all customers and create a new column Transaction_Frequency.

df['Transaction_Frequency'] = df['Number_of_Transactions'].apply(categorize_transactions)
df.head(3)
```

```python
# Create a new column Loan_Eligibility, where:
    # If Credit_Score >= 700 and Account_Balance > 50000 → "Eligible"
    # Otherwise → "Not Eligible"
def eligilibity(loan):
    if loan['Credit_Score'] >= 700 and loan['Account_Balance'] > 50000:
        return 'Eligible'
    else:
         return 'Not Eligible'
df['Loan_Eligibility'] = df.apply(eligilibity, axis=1)
df.head(2)
```

```python
# Question 4:Write a function calculate_debt_to_income_ratio(account_balance, monthly_income) that returns the debt-to-income ratio as (account balance / annual income) * 100.
# defining a function
def calculate_debt_to_income_ratio(account_balance, monthly_income):
    annual_income = monthly_income * 12
    return (account_balance / annual_income) * 100

# o	Apply this function and create a new column Debt_to_Income_Ratio.
df['Debt_to_Income_Ratio'] = df.apply(lambda row: calculate_debt_to_income_ratio(row['Account_Balance'], row['Monthly_Income']), axis=1)

df.head(3)
```

```python
# Question 5: Filter customers with Debt-to-Income Ratio > 50% and display their details.
high_debt_customers = df[df['Debt_to_Income_Ratio'] > 50]
high_debt_customers.head(3)
```

```python
# Write a function to classify customers based on satisfaction:
# 1-2 stars → "Dissatisfied"
# stars → "Neutral"
# 4-5 stars → "Satisfied"

# defining a fucntion
def classify_satisfaction(satisfaction_score):
    if satisfaction_score <= 2:
        return 'Dissatisfied'
    elif satisfaction_score == 3:
        return 'Neutral'
    elif satisfaction_score >= 4:
        return 'Satisfied'

# Create a new column Satisfaction_Category using this function.
df['Satisfaction_Category'] = df['Customer_Satisfaction'].apply(classify_satisfaction)
df.head(2)
```
